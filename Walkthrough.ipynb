## Super Brief Intro

### Original Mechanism

The Goldberger-Wise (GW) mechanism is a method for generating a potential that stabilizes $y_0$ at its minima, thus fixing the size of the extra dimension through a field acquiring an expectation value.  If the mechanism is tuned so that the expected brane seperation is at $ky_0 \backsim 30$ then the mechanism provides an elegant solution to dynamically fitting the graviational heirachy without the need to fine-tuning. The associated potential will simultaneously give the radion mass \cite{RadionGW}. The procedure involves inserting a scalar field into the bulk of the RS model, with the bulk action given by first equation equation below, and the interactions on the hidden and visible branes given by equations second and third equation below respectively

\begin{equation} S_{b}=\frac{1}{2} \int d^{4} x \int_{-y_0}^{y_0} dy \sqrt{G}\left(G^{A B} \partial_{A} \Phi \partial_{B} \Phi-m^{2} \Phi^{2}\right) \end{equation} 
\begin{equation} S_{h}=-\int d^{4} x \sqrt{-g_{h}} \lambda_{h}\left(\Phi^{2}-v_{h}^{2}\right)^{2} \end{equation} 
\begin{equation} S_{v}=-\int d^{4} x \sqrt{-g_{v}} \lambda_{v}\left(\Phi^{2}-v_{v}^{2}\right)^{2} \end{equation} 


$\lambda_h$ and $\lambda_v$ are the hidden and visible brane coupling coefficients respectively; $G_{AB}$ is the 5D spacetime described the line element $ds^2=e^{-2k|y|} \eta_{\mu \nu}d x^{\mu} d x^{\nu}-d y^{2}$; and $g_h$ and $g_v$ are the equivalent 4D metrics at the hidden and visible branes respectively.  Associated with this bulk action, there is also an effective potential, the minima of which dynamically sets the expectation of the brane sepration. 

\begin{equation}\label{eq:GWPotential}
-\int d^{4} x V_{G W}\left(y_{0}\right)=S_b+S_h+S_v
\end{equation}

Returning out attention to the action, it is possible to derive the Euler Lagrange (EL) equation of motion.

\begin{aligned} \partial^2_y\Phi= 4k\partial_y\Phi+m^{2}\Phi  + \lambda_{v} \Phi\left(\Phi^{2}-v_{v}^{2}\right) \delta(y-y_0) +\lambda_{h} \Phi\left(\Phi^{2}-v_{h}^{2}\right) \delta(y) \end{aligned}

Away from the boundaries, the two interaction terms $\lambda_{v} \Phi\left(\Phi^{2}-v_{v}^{2}\right) \delta(y-y_0)$ and $\lambda_{h} \Phi\left(\Phi^{2}-v_{h}^{2}\right) \delta(y)$ do not contribute, which linearises the equation of mation such that its general solution is of the form

\begin{equation} \Phi(y)=e^{2ky}\left[A e^{\nu ky}+B e^{-\nu ky}\right] \end{equation}

Where $\nu=\sqrt{4+m^2/k^2}$. The two unknown coefficients $A$ and $B$ are determined from the boundary conditions, which are obtained in in the original Goldberger-Wise paper by considering the simplified case where the coupling parameters $\lambda_h$ and  $\lambda_v$ are assumed to be large (approximated as infinite). In this case it is energetically favourable for the boundary values to be equal to their respective brane tensions, i.e:  $\Phi(0)=v_{h}$ and $\Phi(y_0)=v_{v}$. Using these "stiff" boundary conditions $A$ and $B$ can be deduced. 

In the case of stiff boundary conditions, it is also possible to deduce that the potential $V_{GW}$ also has an analytical solution, of the form

\begin{equation}\label{eq:GWPotentialA}
\begin{aligned} V_{GW}\left(y_0\right)=& k(\nu+2) A^{2}\left(e^{2 \nu y_0}-1\right)+k(\nu-2) B^{2}\left(1-e^{-2 \nu k y_0}\right) \\ &+\lambda_{v} e^{-4 ky_0}\left(\Phi(y_0)^{2}-v_{v}^{2}\right)^{2}+\lambda_{h}\left(\Phi(0)^{2}-v_{h}^{2}\right)^{2} \end{aligned}
\end{equation}

The minima of the potential forms the expectation for brane sepration, and therefore defines the heirachy. In the large $k_y0$ limit and making the further assumption that $\frac{m}{k} << 1$ the location of the minima $y_-$ can be approximated as.

\begin{equation}\label{eq:MinimaApprox}
\begin{aligned} 
ky_-=\left(\frac{4k^2}{m^2}\right)\ln{\left[\frac{v_h}{v_v}\right]}
\end{aligned}
\end{equation}

Thus, the size of the extra dimension $y_0$ is dynamically set to the position of the potentials minimum $y_-$, which is determined by the four parameters of the potential: $k$, $m$, $v_h$ and $v_v$. Inserting values for these parameters into the equation it is clear that no extreme hierarchy or fine tuning of the parameters is required to create the extra dimensional size that produces the observed hierarchy between the Weak and Planck scales. For example, if $m/k=0.2$ and $v_h/v_v=1.4$ then this yields the required $ky_-$ value of approximately 30.

### Modifications : Finite Coupling Coefficients

However, if the $\lambda_h$ \& $\lambda_v$ values are made finite, then the boundary conditions for $\Phi$ come from the delta functions in the equation of motion. Due to the the requirement of $\Phi$ to be continuous in y, it is the second derivative $\partial^2_y\Phi$ that is proportional to the deltas, meaning that the coefficient of each delta function is proportional to the first derivative $\partial_y\Phi$. Thus, one can deduce the "jump" boundary condition equations.

\begin{equation} [\partial_y\Phi]_{y=0} = \lambda_h \Phi(0)(\Phi(0)^2-\nu^2_h) \end{equation}
    
\begin{equation} [\partial_y\Phi]_{y=y_0} = \lambda_v \Phi(y_0)(\Phi(y_0)^2-\nu^2_v) \end{equation}

Making $\lambda_h$ \& $\lambda_v$ finite will result in "loose" boundary conditions that are dependent on the brane separation $y_0$.


Goldberger and Wise also assume that the field has no bulk quartic self interaction. Introducing this phenomena to the mechanism manifests itself as an extra non localised $\Phi^4$ term in the bulk acton, such that $S_B$ is now given by

\begin{equation}\label{eq:GWBulkActionSelfInt}
S_{b}=\frac{1}{2} \int d^{4} x \int_{-y_0}^{y_0} dy \sqrt{G}\left(G^{A B} \partial_{A} \Phi \partial_{B} \Phi-m^{2} \Phi^{2} - \frac{\lambda_B}{4}\Phi^4\right)
\end{equation}

Iwhere $\lambda_B$ is the bulk self interaction coupling constant, which determines the strength of the self interaction. Introducing a bulk interaction changes Euler-Lagrange equation of motion for the action. Away from the boundaries the equation of motion is now given by

\begin{equation}\label{eq:ELBulk}
\frac{\partial^2\Phi}{\partial y^2}=4k\frac{\partial\Phi}{\partial y} + m^2\Phi +\lambda_B\Phi^3
\end{equation}

The original mechanism makes the assumptions about the coupling of the scalar field for sake of analytical simplicity; addressing these assumptions numerically is the focus of this worksheet. The modified mechanism introduces three extra parameters tunable: $\lambda_h$, $\lambda_v$ \& $\lambda_B$, alongside the 4 hyper parameters of the standard mechanism. The aim of this worksheet is to investigate what impact, if any, values of these extra coupling parameters have on the key featres of both the feild $\Phi(y)$ and the effective potential $V_{GW}(y_0)$

Furthermore, as the size of the extra dimension $y_0$ gets larger, the effects of the bulk interaction term becomes more pronounced: Keeping the bulk coupling $\lambda_B$ constant and increasing the size of the extra dimension appears to have the same effect on the shape of the field as keeping the keeping the size of the extra dimension constant and increasing $\lambda_B$. This is to be expected, as increasing either $y_0$ or $\lambda_B$ will make the $\lambda_B\Phi^3$ term in equation of motion more dominant.

### Effects of finite $\lambda_h$ and $\lambda_v$ on $V_{GW}$

#### Impact on $y_0\xrightarrow{}0$ behaviour of $V_{GW}$
It can be deduced that when the potential $V_{GW}$ is subject to the stiff boundary conditions given by equations, the potential approaches infinite values as $y_0\xrightarrow{}0$. This is an unphysical consequence of stiff boundaries that is removed when the effects of finite boundary coupling are considered. The program above clearly demonstrates that when $\lambda_v$ and $\lambda_h$ are made finite then the divergence of the potential at $y_0=0$ is removed, even if the value of $\lambda_v$ and $\lambda_h$ remain large. As the values of $\lambda_v$ and $\lambda_h$ decrease, the value of $V_{GW}(0)$ also decreases. In  Goldberger and Wise's original paper  it was analytically determined that the value of the potential at the hidden brane, $V_{GW}(0)$, is given by the equation below. This is in direct agreement with the results produced above.

\begin{equation}\label{eq:VGWIntercept}
V_{GW}(0)=\frac{\lambda_{h} \lambda_{v}}{\lambda_{h}+\lambda_{v}}\left(v_{v}^{2}-v_{h}^{2}\right)^{2}
\end{equation}

#### Impact on the rest of $V_{GW}$
Whilst making the $\lambda$ values finite most heavily effects the features of the potential in the vicinity of the hidden brane, the value of the finite $\lambda$ coefficients will also have minor effects on other features of the potential. It can be observed from the model above that the depth of the minima $D=V_{GW}(\infty)-V_{GW}(y_-)$ appears to have a greater dependence on the value of $\lambda_v$ than on $\lambda_h$. Contrariwise, the value of $V_{GW}(\infty)$ is very dependent on $\lambda_h$ but appears to have no dependence on $\lambda_v$. Analytically, it is possible to calculate the leading $1/\lambda$ correction to the potential. This correction is given below

\begin{equation}\label{eq:PotCor}
\delta V_{GW}(0)=-\frac{4k^2}{\lambda_v v_v}e^{-4k y_0}\left(v_{v}-v_{h}e^{-\epsilon k y_0}\right)^{2}
\end{equation}

Using the correction given above, to is possible to produce a $1/\lambda_v$ order approximation for the variation of depth as a function of $\lambda_v$, or $D(\lambda_v)$, given by

\begin{equation}\label{eq:Depthapprox}
D(\lambda_v)=D_s+\frac{4k^2}{\lambda_v v_v}e^{-4k y_0}\left(v_{v}-v_{h}e^{-\epsilon k y_0}\right)^{2} + O(\lambda_v^2, \lambda_h)
\end{equation}

Where $D_s$ is the depth of the minima for the stiff potential. This shows good agreement with the functions above

```python
lams=np.linspace(20, 50, 10)
D=[]
for l in lams :
    print("l: ", l)
    D.append(Bulk_Potential(1.07, 1, 0.1, 3.5, LamH=100, LamV=l, LamB=0, visualise=False, value_returned="depth"))
```

```python
fig=plt.figure(figsize=(12, 6))
plt.plot(1/lams, -np.array(D), '.', linestyle='--')
#plt.xlim(0, 0.1)
#plt.ylim(0, 200)
plt.xlabel("$1/\lambda_v$")
plt.ylabel("$D$")
plt.show()
```

## Radion Mass

The field $\phi(y)$ in the context of the extended 5D Einstein-Hilbert action of $y_0$ dependent fluctuations around the background geometry does not represent the Radion Field, as the kinetic term within the action is not canonically normalized. To evaluate the behaviour of the radion, one must change variables to the canonically normalized radion field $\varphi = fe^{-ky_0}$, where $f=\sqrt{24M_P^3/k}$. The mass of the radion is found by evaluating $\varphi$ excitations about the minimum of the radion potential $V_{GW}$, which is equivalent to the second derivative of the $V_{GW}$ with respect to the radion field evaluated at the minimum $\varphi= \langle \varphi \rangle$. This can also be expressed in terms of the  second derivative of the $V_{GW}$ with respect to $y_0$ at the minimum $y_0=y_-$, as worked out below:

\begin{equation}\nonumber
\begin{split}
    m^2_R&=\left[\frac{d^2 V_{GW}}{d\varphi^2}\right]_{\varphi= \langle \varphi \rangle}=\left[\frac{d}{d\varphi}\left(\frac{dy_0}{d\varphi}\frac{dV_{GW}}{dy_0}\right)\right]_{y_0=y_-}\\ 
    \\
    &= \underbrace{\left[ \frac{d^2y_0}{d\varphi^2}\frac{dV_{GW}}{dy_0}\right]}_{= 0 \text{ as $\frac{dV_{GW}}{dy_0}=0$ at minima}} + \left[ \left(\frac{dy_0}{d\varphi}\right)^2\frac{d^2 V_{GW}}{dy_0^2}\right]_{y_0=y_-}
    = \left[ \left(\frac{dy_0}{d\varphi}\right)^2\frac{d^2 V_{GW}}{dy_0^2}\right]_{y_0=y_-}\\
    \\
    \varphi&=fe^{-ky_0} \xrightarrow{} y_0=-\frac{1}{k}\ln\left({\frac{\varphi}{f}}\right) \xrightarrow{} \frac{dy_0}{d\varphi}=-\frac{1}{\varphi k}\\
    \\
    \therefore m^2_R  &= \left[ \left(-\frac{1}{\varphi k}\right)^2\frac{d^2 V_{GW}}{dy_0^2}\right]_{y_0=y_-} = \frac{e^{2ky_-}}{f^2k^2}\left[\frac{d^2 V_{GW}}{dy_0^2}\right]_{y_0=y_-}\\
\end{split}
\end{equation}

However, this has now introduced an extra free parameter to the numerical analysis, $M$ within the parameter $f$,  which is undesirable. Removing this extra parameter from the numerical analysis requires the use of the analytical expression for Radion mass in the no coupling case, $m_0$. This is given below, where $\epsilon=m^2/4k^2$ and $y_c$ is the minimum of $V_{GW}(y_0)$ with no coupling.
\begin{equation}\label{eq:RadMass0}
m^2_0=\frac{k^2 v_v^2}{3M^3}\epsilon^2 e^{-2ky_c}
\end{equation}

Dividing equation these two expressions, one obtains an expression for the ratio between the squared radion masses with and without coupling. This expression is given below, and is independent of the parameter $M$.

\begin{equation}\label{eq:RadMassRatio}
\left(\frac{m_R}{m_0} \right)^2= \frac{e^{2k(y_- + y_c)}}{8k^3\epsilon^2 v_v^2} \left[\frac{d^2 V_{GW}}{dy_0^2}\right]_{y_0=y_-}
\end{equation}

The equation above allows the dimensionless ratio $m_R/m_0$ to be calculated from the second derivative of the GW potential with respect to $y_0$  which can be determined numerically without the need of any extra parameters. This expresion is very senstive to numerical instability, as the first term is a large value, and the second term is a very small value which multiply together

```python
lams=np.linspace(0, 0.1, 10)
m=[]
for l in lams :   
    m.append(Bulk_Potential(1.07, 1, 0.1, 3.5, LamB=l, visualise=False, value_returned="mass"))
```

```python
fig=plt.figure(figsize=(12, 6))
plt.plot(lams, m, '.')
#plt.xlim(0, 0.1)
#plt.ylim(0, 200)
plt.xlabel("$\lambda_B$")
plt.ylabel("$m_r^2/m_0^2$")
plt.show()
fig=plt.figure(figsize=(12, 6))
plt.plot(lams, np.sqrt(m), '.')
#plt.xlim(0, 0.1)
#plt.ylim(0, 200)
plt.xlabel("$\lambda_B$")
plt.ylabel("$m_r/m_0$")
plt.show()
```

_Above graph shows the variation of relative squared mass (first) and realtive mass (second) compared to the uncoupled limit $1/\lambda_B$.

## Further Extensions to Current Program

### Back reaction
Throughout we have considered how the bulk scalar propagates within the Randall-Sundrum geometry, without considering how the presence of the bulk scalar warps this geometry through back-reaction. The effects of back-reaction in the mechanism as originally outlined are understood to be fairly small, and can be taken account for via the Superpotential method. However, the methods used in this program do not account for the effects of back reaction, thus there is likely a $\lambda_B$ cutoff at which the mechanism without back reaction becomes an invalid approximation.  

Finding the value of this $\lambda_B$ cutoff would be an additional useful result. This could be done by calculating and comparing the Stress-Energy Tensors due to the the scalar field and the bulk cosmological constant. The limit at which the stress-energy tensor due to the the scalar field can no longer be neglected represents the cutoff at which the back-reaction must be considered. Better still, the programs used in this thesis could be made more complex to include the Einstein equations in the system of differential equations; the resulting numerical solution to this system equations of motions would therefore represent a full solution including back-reaction.

### Realistic Hierarchies
The hierarchy that would be produced by the parameters in this program does not represent the hierarchy that exists in nature, which would produced by a brane seperation of $ky_0\approx30$ in the Randall-Sundrum setup. Whilst the results of this this program on small heirarchies still provide an insight into how the mechanism is effected by variation of the coupling coefficients, it would be useful to workout the extent of which these effects manifest themselves in realistic hierarchies. This could be done by either somehow imporving the program to deal with numerics assocaited withlarge hierarchies, or by deducing how the $\lambda$ dependencies vary with the remaining parameters $k$, $m$, $v_h$ \& $v_v$.
